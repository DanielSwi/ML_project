```python
print(f'Size of the data before outliers removal{df.shape}')
cols = ['total_duration', 'info_page_duration','product_page_duration', 'admin_page_duration'] # one or more

Q1 = df[cols].quantile(0.1)
Q3 = df[cols].quantile(0.9)
IQR = Q3 - Q1

df = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
print(f'Size of the data after outliers removal{df.shape}')
```

From the first table printed, we can see that the features regarding the connections in each page and their duration as well as the total duration (spacially this one), have a lot of Nan valaues.     
I chose to fill them (the nan's) in the following way:  
    -Each page category will be filled by the mediant of each feature.  
    -Each time_duration will be fill by the mean of the feature only if the number of page connections (for the same page and in the same row), in not 0.                               
    -Each total_duration will be fill by the sum of the 3 other time features.

```python
#Filing the duration features

df["num_of_admin_pages"] = df["num_of_admin_pages"].fillna(df["num_of_admin_pages"].median())
df["num_of_info_pages"] = df["num_of_info_pages"].fillna(df["num_of_info_pages"].median())
df["num_of_product_pages"] = df["num_of_product_pages"].fillna(df["num_of_product_pages"].median())


df.loc[df["num_of_admin_pages"] > 0,"admin_page_duration"] = df.loc[df["num_of_admin_pages"] > 0,"admin_page_duration"].fillna(df['admin_page_duration'].mean())
df.loc[df["num_of_admin_pages"] == 0,"admin_page_duration"] = df.loc[df["num_of_admin_pages"] == 0,"admin_page_duration"].fillna(0)

df.loc[df["num_of_info_pages"] > 0,"info_page_duration"] = df.loc[df["num_of_info_pages"] > 0,"info_page_duration"].fillna(df['info_page_duration'].mean())
df.loc[df["num_of_info_pages"] == 0,"info_page_duration"] = df.loc[df["num_of_info_pages"] == 0,"info_page_duration"].fillna(0)

df.loc[df["num_of_product_pages"] > 0,"product_page_duration"] = df.loc[df["num_of_product_pages"] > 0,"product_page_duration"].fillna(df['product_page_duration'].mean())
df.loc[df["num_of_product_pages"] == 0,"product_page_duration"] = df.loc[df["num_of_product_pages"] == 0,"product_page_duration"].fillna(0)

#Filing total duration feature

durations_lst = ["info_page_duration","admin_page_duration","product_page_duration"]

df['total_duration'] = df['total_duration'].fillna( df[durations_lst].sum(axis = 1))
```

Now let's fill the rest of the missing values :

```python
#Replacing Na's in BounceRates, Exit Rates,Page Values and closeness to holiday by mean value in each column

df["BounceRates"] = df["BounceRates"].fillna(df["BounceRates"].median())
df["ExitRates"] = df["ExitRates"].fillna(df["ExitRates"].median())
df["PageValues"] = df["PageValues"].fillna(df["PageValues"].median())
df["closeness_to_holiday"] = df["closeness_to_holiday"].fillna(df["closeness_to_holiday"].median())

#Filling B with mean value of the feature
df["B"] = df["B"].fillna(df["B"].mean())
```

```python
# Multi Layer Perceptron Artificial Neural Network
from sklearn.neural_network import MLPClassifier    
mlpc_params = {"alpha": [0.1, 0.001],
              "hidden_layer_sizes": [(10,10,10),
                                     (100,100,100),
                                     (100,100)],
              "solver" : ["lbfgs","sgd"],
              "activation": ["relu","logistic"]}





mlpc = MLPClassifier(random_state = 0) # ANN model object created

# Model CV process 
mlpc_cv_model = GridSearchCV(mlpc, mlpc_params, 
                         cv = 3, # To make a 3-fold CV
                         n_jobs = -1, # Number of jobs to be run in parallel (-1: means to use all processors)
                         verbose = 2) # Controls the level of detail: higher means more messages gets value as integer.

mlpc_cv_model.fit(X_train, y_train) 


# The best parameter obtained as a result of CV process

print("The best parameters: " + str(mlpc_cv_model.best_params_))
# Model Tuning
# Setting the Final Model with the best parameter

mlpc_tuned = mlpc_cv_model.best_estimator_

# Fitting Final Model
mlpc_tuned.fit(X_train, y_train)
```

```python
from sklearn.model_selection import StratifiedKFold

# Run classifier with cross-validation and plot ROC curves
cv = StratifiedKFold(n_splits=5)

tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

fig, ax = plt.subplots()
for i, (train, test) in enumerate(cv.split(X, y)):
    mlpc_tuned.fit(X.iloc[train], y.iloc[train])
    viz = RocCurveDisplay.from_estimator(
        mlpc_tuned,
        X.iloc[test],
        y.iloc[test],
        name="ROC fold {}".format(i),
        alpha=0.3,
        lw=1,
        ax=ax,
    )
    interp_tpr = np.interp(mean_fpr, viz.fpr, viz.tpr)
    interp_tpr[0] = 0.0
    tprs.append(interp_tpr)
    aucs.append(viz.roc_auc)

ax.plot([0, 1], [0, 1], linestyle="--", lw=2, color="r", label="Chance", alpha=0.8)

mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = np.mean(aucs)
std_auc = np.std(aucs)
ax.plot(
    mean_fpr,
    mean_tpr,
    color="b",
    label=r"Mean ROC (AUC = %0.2f $\pm$ %0.2f)" % (mean_auc, std_auc),
    lw=2,
    alpha=0.8,
)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
ax.fill_between(
    mean_fpr,
    tprs_lower,
    tprs_upper,
    color="grey",
    alpha=0.2,
    label=r"$\pm$ 1 std. dev.",
)

ax.set(
    xlim=[-0.05, 1.05],
    ylim=[-0.05, 1.05],
    title="Receiver operating characteristic Multi-Layer Perceptron",
)
ax.legend(loc="lower right")
plt.show()
```
